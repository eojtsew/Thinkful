```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
hate_crimes = pd.read_csv('hate_crime_data.csv')
```

```python
hate_crimes['Total'] = hate_crimes.iloc[:,10:14].sum(axis=1)
hate_crimes = hate_crimes.dropna()
```

```python
plt.figure(figsize=(15,5))
plt.scatter(hate_crimes.Population,hate_crimes.Total, alpha=0.5)
plt.axis([100,8500000,0,350])
plt.xlabel('Population Size')
plt.ylabel('Hate Crimes')
plt.title('Hate Crimes Reported in the US in 2013 vs Population Size (by Precinct)')
plt.show()
```

The figure above shows the number of hate crimes reported per precinct versus the precinct's population size.

```python
states = hate_crimes.groupby('State').Total.sum().reset_index()
state_pops = hate_crimes.groupby('State').Population.sum().reset_index()
states = states.merge(state_pops)
```

```python
plt.scatter(states.Population, states.Total)
plt.xlabel('Population Size')
plt.ylabel('Hate Crimes')
plt.title('Hate Crimes Reported in the US in 2013 vs Population Size (by state)')
plt.show()
```

The figure above shows the number of hate crimes reported by state versus the state's population size. 

```python
plt.figure(figsize=(15,5))
plt.scatter(states.Population,states.Total, alpha=0.5)
plt.xlabel('Population Size')
plt.ylabel('Hate Crimes')
plt.title('Hate Crimes Reported in the US in 2013 vs Population Size (by state)')
plt.show()
```
