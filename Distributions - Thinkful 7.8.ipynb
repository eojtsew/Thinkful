```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Binomial Distribution
binomialvar = np.random.binomial(n=45, p=0.5, size=100)

plt.hist(binomialvar, bins=20, color='c')

# Add a vertical line at the mean.
plt.axvline(binomialvar.mean(), color='b', linestyle='solid', linewidth=2)

# Add a vertical line at one standard deviation above the mean.
plt.axvline(binomialvar.mean() + binomialvar.std(), color='b', linestyle='dashed', linewidth=2)

# Add a vertical line at one standard deviation below the mean.
plt.axvline(binomialvar.mean()-binomialvar.std(), color='b', linestyle='dashed', linewidth=2) 

# Print the histogram.
plt.show()
```

```python
# Exponential Distribution
exponentialvar = np.random.exponential(scale=1.5, size=100)

plt.hist(exponentialvar, bins=20, color='c')

plt.axvline(exponentialvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(exponentialvar.mean() + exponentialvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(exponentialvar.mean()-exponentialvar.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

```python
# Gamma Distribution
gammavar = np.random.gamma(shape=2.0, scale=1.5, size=100)

plt.hist(gammavar, bins=20, color='c')

plt.axvline(gammavar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(gammavar.mean() + gammavar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(gammavar.mean()-gammavar.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

```python
# Hypergeometric Distribution
hypergeometricvar = np.random.hypergeometric(ngood=45, nbad=45, nsample=45, size=100)

plt.hist(hypergeometricvar, bins=20, color='c')

plt.axvline(hypergeometricvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(hypergeometricvar.mean() + hypergeometricvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(hypergeometricvar.mean()-hypergeometricvar.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

```python
# Poisson Distribution
poissonvar = np.random.poisson(lam=1.5, size=100)

plt.hist(poissonvar, bins=20, color='c')

plt.axvline(poissonvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(poissonvar.mean() + poissonvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(poissonvar.mean()-poissonvar.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```

```python
# Logistic Distribution
logisticvar = np.random.logistic(loc=2.0, scale=1.5, size=100)

plt.hist(logisticvar, bins=20, color='c')

plt.axvline(logisticvar.mean(), color='b', linestyle='solid', linewidth=2)
plt.axvline(logisticvar.mean() + logisticvar.std(), color='b', linestyle='dashed', linewidth=2)
plt.axvline(logisticvar.mean()-logisticvar.std(), color='b', linestyle='dashed', linewidth=2) 

plt.show()
```
