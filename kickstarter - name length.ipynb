```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
kicks = pd.read_csv("kickstarter-projects/ks-projects-201801.csv")
```

```python
kicks['name'] = list(map(str,kicks['name']))
kicks['name_length'] = list(map(len,kicks['name']))
```

```python
lengths_means = kicks.groupby('name_length').backers.mean()
lengths_medians = kicks.groupby('name_length').backers.median()
length_counts = kicks.groupby('name_length').ID.count()
```

```python
plt.bar(length_counts.index,length_counts)
plt.show()
```

```python
plt.scatter(lengths_means.index,lengths_means)
plt.show()
```

```python
plt.scatter(lengths_medians.index,lengths_medians)
plt.show()
```

```python
kicks_win = kicks[kicks.state=='successful']
kicks_fail = kicks[kicks.state=='failed']
plt.hist(kicks_fail.name_length,density=True,bins=85,color='red',alpha=0.5)
plt.hist(kicks_win.name_length,density=True,bins=85,color='blue',alpha=0.5)
plt.show()
```
