```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('football_results.csv')
```

```python
data['abs_margin'] = abs(data['home_score'] - data['away_score'])
data['margin'] = data['home_score'] - data['away_score']
```

```python
def win(margin):
    if margin > 0:
        return 'home'
    if margin < 0:
        return 'away'
    if margin == 0:
        return 'tie'

data['winner'] = data.margin.apply(win)
data.head()
```

```python
# use seaborn defaults
sns.set(style='white')

# histogram
sns.distplot(data.margin, kde=False)
plt.show()

# histogram w/rug
sns.distplot(data.margin, kde=False, rug=True)
plt.show()

# boxplot
ax = sns.boxplot(y='margin',data=data)  
plt.title('Final Score Margin')
sns.despine(offset=10, trim=True)
ax.set(xlabel='', ylabel='Points')
plt.show()

#kdeplot
sns.kdeplot(data.margin, shade=True)
plt.show()

# neutral or not boxplots
sns.boxplot(x='neutral',y='margin',data=data)
plt.show()

# home or away boxplots
sns.boxplot(x='winner',y='abs_margin',data=data[data.neutral == False])
plt.show()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
# scatterplot with regression line
ax = sns.lmplot(y='sepal_length',x='sepal_width',data=iris,col='species',fit_reg=True,scatter_kws={'alpha':0.4})
ax.set_ylabels('Sepal Length')
ax.set_xlabels('Sepal Width')
plt.title('Sepal Length vs Sepal Width')
plt.show()
```

```python
fish = pd.read_csv('fish.csv')
fish.head()
```

```python
# joint plot of fish height and weight
sns.jointplot(x='Height',y='Weight',data=fish)
plt.show()
```

```python
# hexbin plot of fish height and weight
sns.jointplot(x='Height',y='Weight',data=fish, kind='hex')
plt.show()
```

```python
# relplot
sns.relplot(x='Height',y='Weight',hue='Species',data=fish)
plt.show()
```

```python
# pairplot
sns.pairplot(fish, hue='Species')
plt.show()
```
